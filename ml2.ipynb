```python
import torch.nn as nn
import torch.nn.functional as F
```

```python
class OurNet(nn.Module):

    def __init__(self):
        super(OurNet, self).__init__()
        self.fc1 = nn.Linear(3, 200)
        self.fc2 = nn.Linear(200, 200)
        self.fc3 = nn.Linear(200, 2)


    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = F.relu(self.fc2(x))
        x = self.fc3(x)
        return F.softmax(F.normalize(x.abs()))
```

```python
loss_fun = nn.BCELoss()
optimizer = torch.optim.SGD(net.parameters(), lr=0.01)
```

```python
def get_data(name):
    train = pd.read_csv(name)
    train = train[["pclass", "age", "sex", 'survived']]
    median = train['age'].median()
    train['age'].fillna(median, inplace=True)
    train['pclass'].replace('3rd', 3, inplace=True)
    train['pclass'].replace('2nd', 2, inplace=True)
    train['pclass'].replace('1st', 1, inplace=True)
    train['sex'] = np.where(train['sex'] == 'female', 0, 1)
    return train
train_data = get_data("data/titanic_train.csv")
test_data = get_data("data/titanic_test.csv")
```

```python
X = torch.Tensor(train_data[["pclass", "age", "sex"]].to_numpy())
Y = torch.Tensor(train_data[["survived"]].to_numpy())
```

```python
Y_new = torch.Tensor(len(Y), 2)
for i in range(len(Y)):
    print(Y[i])
    if np.abs(Y[i]) < 1.e-10:
        Y_new[i][0] = 1
        Y_new[i][1] = 0
    else:
        Y_new[i][0] = 0
        Y_new[i][1] = 1
Y = Y_new
```

```python
net = OurNet()
loss_fun = nn.BCELoss()
optimizer = torch.optim.SGD(net.parameters(), lr=0.01)
```

```python
epochs = 10_000
batch_size = 10

def fit(net, batch_size, X, Y):
    net.train(True)
    sumL, sumA, numB = 0, 0, int(len(X) / batch_size)

    for i in range(0, numB*batch_size, batch_size):
        xb = X[i: i + batch_size]
        yb = Y[i: i + batch_size]

        y_cur = net(xb)
        L = loss_fun(y_cur, yb)

        optimizer.zero_grad()
        L.backward()
        optimizer.step()

        sumL += L.item()
    return sumL

for epoch in range(epochs):
    ls = fit(net,batch_size, X, Y)
    if epoch % 100 == 0:
        print(f"{epoch}\t:\tloss = {ls}")
```
